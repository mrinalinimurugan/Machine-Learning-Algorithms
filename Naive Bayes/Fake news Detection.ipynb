# Read data

# Create new colomn to determine fake/true news

```python
truenews['True/Fake']='True'
fakenews['True/Fake']='Fake'
```

# Concatenate true anf fake news and add column article 

```python
news=pd.concat([truenews,fakenews])
news['Article']=news['title']+news['text']
news.sample(frac=1)
```

# Preprocessing of data

```python
from nltk.corpus import stopwords
import string
```

```python
def processtxt(l):
    #check string if they have punctuation
    nopunc=[char for char in l if char not in string.punctuation]
    #Join it again to form a string 
    nopunc=''.join(nopunc)
    #To convert string to lowercase and remove stopwords(very commony used words)
    cleanstr=[word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    return cleanstr
```

```python
#Tokenize the article
news['Clean txt'] = news['Article'].apply(processtxt)
```

```python
news.sample(10)
```

```python
#bow=bag of words
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(analyzer=processtxt).fit(news['Clean txt'])
print(len(bow_transformer.vocabulary_))
```

```python
news_bow=bow_transformer.transform(news['Clean txt'])
```

```python
#sparsity refers to a matrix of numbers that includes many zeros or values that will not significantly impact a calculation.
print("shape of sparse matrix:",news_bow.shape)
print("No of non zero occurances:",news_bow.nnz)
```

```python
sparsity=(100*news_bow.nnz/(news_bow.shape[0]*news_bow.shape[1]))
print('sparsity{}:'.format(round(sparsity)))
          
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidfittrans = TfidfTransformer().fit(news_bow)
news_tfid = tfidfittrans.transform(news_bow)
print(news_tfid.shape)
```

# Train the model using Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
fakenews_detect_model = MultinomialNB().fit(news_tfid,news['True/Fake'])
```

```python
#Model Evaluation
newspredictions=fakenews_detect_model.predict(news_tfid)
print(newspredictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(news['True/Fake'],newspredictions))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(news['Article'],news['True/Fake'],test_size=0.35)
```

```python
from sklearn.pipeline import Pipeline
pl = Pipeline([
    ('bow', CountVectorizer(analyzer=processtxt)),
    ('tfidf', TfidfTransformer()),
    ('classifier',MultinomialNB())
])
pl.fit(xtrain,ytrain)
```

```python
pred = pl.predict(xtest)
```

```python
print(classification_report(pred,ytest))
```
